```python
## Two Sum problem

# Two Sum : Check if a pair with given sum exists in Array



# Problem Statement: Given an array of integers arr[] and an integer target.

# 1st variant: Return YES if there exist two numbers such that their sum is equal to the target. Otherwise, return NO.

# 2nd variant: Return indices of the two numbers such that their sum is equal to the target. Otherwise, we will return {-1, -1}.

# Input: N = 5, arr[] = {2,6,5,8,11}, target = 14
# Output : YES
# Explanation: arr[1] + arr[3] = 14. So, the answer is “YES” for first variant for second variant output will be : [1,3].

# Input: N = 5, arr[] = {2,6,5,8,11}, target = 15
# Output : NO.
# Explanation: There exist no such two numbers whose sum is equal to the target. 

```

```python
# Variant - 1 :Return YES/NO if such a pair exists or not (Brute Force Approach)
def two_sum_exists_brute_force(N, arr, target):
    for i in range(len(arr)):
        current_number = arr[i]
        for j in range(i + 1, len(arr)):
            if current_number + arr[j] == target:
                return "YES"
    return "NO" 

# Variant - 2 : Return indices of the two numbers such that their sum is equal to the target (Brute Force Approach)
def two_sum_indices_brute_force(N, arr, target):
    for i in range(len(arr)):
        current_number = arr[i]
        for j in range(i + 1, len(arr)):
            if current_number + arr[j] == target:
                return [i, j]
    return [-1, -1]
```

```python
print(two_sum_exists_brute_force(N=5, arr=[2,6,5,8,11], target=14))  # Output: YES
print(two_sum_exists_brute_force(N=5, arr=[2,6,5,8,11], target=15))  # Output: NO
```

```python
## Better Solution -> Using a hashmap
def two_sum_hashmap(N, arr, target):
    map ={}
    for i in range(0, len(arr)):
        current_number = arr[i]
        complement = target -current_number
        if complement in map:
            return "YES"        # Variant-1
            # return [i, map[complement]      # Variant-2
        else:
            map[current_number] = i
        # print(map, arr[i])
    return "NO"
    # return [-1, -1]       # Variant-2
print(two_sum_hashmap(N=5, arr=[2,6,5,8,11], target=14))  # Output: YES
print(two_sum_hashmap(N=5, arr=[2,6,5,8,11], target=15))  # Output: NO     
```

```python
# Optimal Solution - > Two Pointer

def two_sum_2_pointer(N, arr, target):
    # Create list of tuples (value, original_index)
    nums_with_index = [(num, idx) for idx, num in enumerate(arr)]
        
    # Sort the list by values
    nums_with_index.sort(key=lambda x: x[0])

    left, right = 0, len(arr) - 1
    
    while left < right:
        current_sum = nums_with_index[left][0] + nums_with_index[right][0]
        if current_sum == target:
            # Return original indices of found elements
            return [nums_with_index[left][1], nums_with_index[right][1]]
        elif current_sum < target:
            # Move left pointer right to increase sum
            left += 1
        else:
            # Move right pointer left to decrease sum
            right -= 1
    
    # No valid pair found
    return [-1, -1]
print(two_sum_2_pointer(N=5, arr=[2,6,5,8,11], target=14))  # Output: YES
print(two_sum_2_pointer(N=5, arr=[2,6,5,8,11], target=15))  # Output: NO      
```
