## Conceptual Notes: The Greedy Choice

### Why Greedy Works
In problems like **Container With Most Water**, the area is limited by the *shorter* line.

**Key Insight:**
```
Area = min(height[left], height[right]) Ã— (right - left)
```

**Greedy Logic:**
- If `height[left] < height[right]`, then `left` is the bottleneck.
- Moving `right` inward CANNOT increase area because:
  - Width decreases
  - Height is still limited by `left`
- Therefore, we MUST move `left` to have any chance of improvement.

This **elimination argument** proves we don't miss the optimal solution!

## Problem 1: Container With Most Water â­â­ Medium
**Task**: You are given an integer array `height` of length `n`. There are `n` vertical lines. Find two lines that together with the x-axis form a container, such that the container contains the most water.

**Visualization:**
```
  |         |      
  |    |    |  |   
  |    |    |  |   
__|____|____|__|___
  L              R
```

````python
def maxArea(height: list[int]) -> int:
    """
    Algorithm:
    1. Start with widest container (left=0, right=n-1)
    2. Calculate area = min(height[left], height[right]) * (right - left)
    3. Greedy: Move the shorter line inward (it's the bottleneck)
    4. Repeat until pointers meet
    
    Time: O(N), Space: O(1)
    """
    left = 0
    right = len(height) - 1
    max_water = 0
    
    while left < right:
        # TODO: Calculate current width
        width = right - left
        
        # TODO: Calculate current area (use min height)
        current_area = min(height[left], height[right]) * width
        
        # TODO: Update max_water
        max_water = max(max_water, current_area)
        
        # TODO: Greedy Move - move the shorter line
        # if height[left] < height[right]:
        #     left += 1
        # else:
        #     right -= 1
        pass
        
    return max_water
````

````python
# Test Cases
assert maxArea([1,8,6,2,5,4,8,3,7]) == 49  # Between index 1 (h=8) and 8 (h=7)
assert maxArea([1,1]) == 1
assert maxArea([4,3,2,1,4]) == 16  # Between index 0 and 4

print("Container With Most Water Tests Passed! âœ“")
````

## Problem 2: Trapping Rain Water â­â­â­ Hard
**Task**: Given `n` non-negative integers representing an elevation map where the width of each bar is 1, compute how much water it can trap after raining.

**Key Insight**: Water at position `i` = `min(max_left, max_right) - height[i]`

**Approaches:**
1. **DP**: Pre-compute max_left[] and max_right[] â†’ O(N) space
2. **Two Pointers**: Track running max from each side â†’ O(1) space

**Why Two Pointers Work:**
- We only care about the SMALLER of the two max boundaries
- If `left_max < right_max`, water at left is determined by left_max
- We don't need to know exact right_max, just that it's bigger!

````python
# First, let's understand with the DP approach
def trap_dp(height: list[int]) -> int:
    """DP approach - O(N) space - for understanding."""
    if not height:
        return 0
    
    n = len(height)
    
    # TODO: Compute max height to the LEFT of each position
    max_left = [0] * n
    max_left[0] = height[0]
    for i in range(1, n):
        max_left[i] = max(max_left[i-1], height[i])
    
    # TODO: Compute max height to the RIGHT of each position
    max_right = [0] * n
    max_right[n-1] = height[n-1]
    for i in range(n-2, -1, -1):
        max_right[i] = max(max_right[i+1], height[i])
    
    # TODO: Calculate water at each position
    water = 0
    for i in range(n):
        water += min(max_left[i], max_right[i]) - height[i]
    
    return water

# Test DP approach first
assert trap_dp([0,1,0,2,1,0,1,3,2,1,2,1]) == 6
print("DP approach works! âœ“")
````

````python
def trap(height: list[int]) -> int:
    """
    Two Pointer approach - O(1) space.
    
    Algorithm:
    1. Track left_max and right_max as we go
    2. Process the side with smaller max (it's the bottleneck)
    3. Water = smaller_max - current_height
    """
    if not height:
        return 0
    
    left, right = 0, len(height) - 1
    left_max, right_max = height[left], height[right]
    water = 0
    
    while left < right:
        if left_max < right_max:
            # Left side is the bottleneck
            left += 1
            
            # TODO: Update left_max
            # left_max = max(left_max, height[left])
            
            # TODO: Add water trapped at this position
            # water += left_max - height[left]
            pass
            
        else:
            # Right side is the bottleneck
            right -= 1
            
            # TODO: Update right_max
            # right_max = max(right_max, height[right])
            
            # TODO: Add water trapped at this position
            # water += right_max - height[right]
            pass
            
    return water
````

````python
# Test Cases
assert trap([0,1,0,2,1,0,1,3,2,1,2,1]) == 6
assert trap([4,2,0,3,2,5]) == 9
assert trap([]) == 0
assert trap([1]) == 0

print("Trapping Rain Water Tests Passed! âœ“")
````
