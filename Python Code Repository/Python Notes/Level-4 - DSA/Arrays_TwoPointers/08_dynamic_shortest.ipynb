- **Shortest**: Shrink when VALID

### Pattern
```python
for right: EXPAND
    while VALID():      # While condition MET
        update_min()    # Current works!

```

## Problem 1: Minimum Size Subarray Sum â­â­ Medium

````python
def minSubArrayLen(target: int, nums: list[int]) -> int:
    left = 0
    current_sum = 0
    min_len = float('inf')
    
    for right in range(len(nums)):
        current_sum += nums[right]
        
        while current_sum >= target:
            min_len = min(min_len, right - left + 1)
            current_sum -= nums[left]
            left += 1
    
    return 0 if min_len == float('inf') else min_len

# Tests
assert minSubArrayLen(7, [2,3,1,2,4,3]) == 2
assert minSubArrayLen(4, [1,4,4]) == 1
print("Min Size Subarray Sum Tests Passed! âœ“")
````

## Problem 2: Minimum Window Substring â­â­â­ Hard

````python
from collections import Counter

def minWindow(s: str, t: str) -> str:
    if not t or not s: return ""
    
    target = Counter(t)
    required = len(target)
    window = {}
    formed = 0
    left = 0
    min_len = float('inf')
    result = ""
    
    for right in range(len(s)):
        char = s[right]
        window[char] = window.get(char, 0) + 1
        
        if char in target and window[char] == target[char]:
            formed += 1
        
        while formed == required:
            if right - left + 1 < min_len:
                min_len = right - left + 1
                result = s[left:right + 1]
            
            left_char = s[left]
            window[left_char] -= 1
            if left_char in target and window[left_char] < target[left_char]:
                formed -= 1
            left += 1
    
    return result

# Tests
assert minWindow("ADOBECODEBANC", "ABC") == "BANC"
print("Min Window Substring Tests Passed! âœ“")
````

## Problem 3: Minimum Window Subsequence â­â­â­ Hard

````python
def minWindowSubsequence(s1: str, s2: str) -> str:
    m, n = len(s1), len(s2)
    min_len = float('inf')
    result = ""
    i = 0
    
    while i < m:
        j = 0
        while i < m and j < n:
            if s1[i] == s2[j]: j += 1
            i += 1
        
        if j < n: break
        end = i - 1
        
        j = n - 1
        while j >= 0:
            if s1[end] == s2[j]: j -= 1
            end -= 1
        
        start = end + 1
        if i - start < min_len:
            min_len = i - start
            result = s1[start:i]
        i = start + 1
    
    return result

# Tests
assert minWindowSubsequence("abcdebdde", "bde") == "bcde"
print("Min Window Subsequence Tests Passed! âœ“")
````
