## Conceptual Notes: The Tournament

When merging two sorted arrays, think of a tournament:
- Compare the "champions" (current elements) of both arrays
- The winner (smaller/larger) gets placed
- Advance that array's pointer

### Merge from End Trick
When merging INTO an existing array:
- Filling from FRONT overwrites unprocessed data!
- Fill from BACK into empty space
```
[1, 2, 3, 0, 0, 0]  <- nums1 with extra space
             ^  ^  Start filling here!
```

## Problem 1: Merge Sorted Array â­ Easy
**Task**: Merge `nums2` into `nums1`. nums1 has length `m + n` with last n elements as 0.

````python
def merge(nums1: list[int], m: int, nums2: list[int], n: int) -> None:
    """
    Algorithm:
    1. Start from END of both arrays
    2. Place LARGER element at the back of nums1
    3. Move corresponding pointer
    """
    p1 = m - 1      # Last valid element in nums1
    p2 = n - 1      # Last element in nums2
    write = m + n - 1  # Last position in nums1
    
    # While both arrays have elements
    while p1 >= 0 and p2 >= 0:
        if nums1[p1] > nums2[p2]:
            nums1[write] = nums1[p1]
            p1 -= 1
        else:
            nums1[write] = nums2[p2]
            p2 -= 1
        write -= 1
    
    # TODO: Copy remaining elements from nums2 (if any)
    # Note: We don't need to copy remaining nums1 - they're already in place!
    while p2 >= 0:
        nums1[write] = nums2[p2]
        p2 -= 1
        write -= 1
````

````python
# Test Cases
n1 = [1,2,3,0,0,0]
merge(n1, 3, [2,5,6], 3)
assert n1 == [1,2,2,3,5,6]

n2 = [1]
merge(n2, 1, [], 0)
assert n2 == [1]

n3 = [0]
merge(n3, 0, [1], 1)
assert n3 == [1]

print("Merge Sorted Array Tests Passed! âœ“")
````

## Problem 2: Squares of a Sorted Array â­ Easy
**Task**: Return squares of each number, sorted.

**Insight**: Largest squares are at the ENDS (most negative or most positive)!

````python
def sortedSquares(nums: list[int]) -> list[int]:
    """
    Algorithm:
    1. Two pointers at ends
    2. Compare absolute values
    3. Square the larger and place at END of result
    """
    n = len(nums)
    result = [0] * n
    left, right = 0, n - 1
    
    for i in range(n - 1, -1, -1):  # Fill from back
        if abs(nums[left]) > abs(nums[right]):
            result[i] = nums[left] ** 2
            left += 1
        else:
            result[i] = nums[right] ** 2
            right -= 1
    
    return result
````

````python
# Test Cases
assert sortedSquares([-4,-1,0,3,10]) == [0,1,9,16,100]
assert sortedSquares([-7,-3,2,3,11]) == [4,9,9,49,121]
assert sortedSquares([1]) == [1]

print("Squares of Sorted Array Tests Passed! âœ“")
````

## Problem 3: Intersection of Two Arrays â­ Easy
**Task**: Return unique elements present in both arrays.

````python
def intersection(nums1: list[int], nums2: list[int]) -> list[int]:
    """Two pointer approach after sorting."""
    nums1.sort()
    nums2.sort()
    
    p1, p2 = 0, 0
    result = set()  # Set for uniqueness
    
    while p1 < len(nums1) and p2 < len(nums2):
        if nums1[p1] == nums2[p2]:
            result.add(nums1[p1])
            p1 += 1
            p2 += 1
        elif nums1[p1] < nums2[p2]:
            p1 += 1
        else:
            p2 += 1
    
    return list(result)
````

````python
# Test Cases
assert sorted(intersection([1,2,2,1], [2,2])) == [2]
assert sorted(intersection([4,9,5], [9,4,9,8,4])) == [4,9]

print("Intersection Tests Passed! âœ“")
````

**Challenge**: Return intersection with duplicates (as many times as they appear in both).

````python
def intersect(nums1: list[int], nums2: list[int]) -> list[int]:
    """Include duplicates in result."""
    nums1.sort()
    nums2.sort()
    
    p1, p2 = 0, 0
    result = []  # List, not set!
    
    while p1 < len(nums1) and p2 < len(nums2):
        # TODO: Implement similar logic but append to list
        if nums1[p1] == nums2[p2]:
            result.append(nums1[p1])
            p1 += 1
            p2 += 1
        elif nums1[p1] < nums2[p2]:
            p1 += 1
        else:
            p2 += 1
    
    return result

# Test
assert intersect([1,2,2,1], [2,2]) == [2,2]
print("Intersection II Passed! âœ“")
````
