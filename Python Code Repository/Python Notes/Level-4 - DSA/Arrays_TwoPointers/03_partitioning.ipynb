## Conceptual Notes: The "Snowball" Method

### The Pattern
Instead of pointers moving towards each other, they move in the SAME direction:
- **Slow Pointer (Write)**: Where to place the next valid element
- **Fast Pointer (Read)**: Scans through the array

**Visualization:**
```
[ 1, 3, 12 | 0, 0 | 5, ... ]
            ^      ^
         write   read
   [valid]  [garbage] [unseen]
```

**Invariant**: Everything before `write` is valid. Between `write` and `read` is garbage.

### When to Use
- Removing elements in-place
- Moving elements to one side
- Partitioning by condition

## ðŸ”¥ Warm-Up: Remove Element
**Goal**: Understand the read/write pointer paradigm.

**Task**: Remove all instances of `val` from array in-place. Return new length.

````python
def removeElement(nums: list[int], val: int) -> int:
    """Remove all occurrences of val in-place."""
    write = 0
    
    for read in range(len(nums)):
        # TODO: If current element is NOT val, keep it
        if nums[read] != val:
            nums[write] = nums[read]
            write += 1
    
    return write

# Test
arr = [3, 2, 2, 3]
k = removeElement(arr, 3)
assert k == 2 and sorted(arr[:k]) == [2, 2]
print("Warm-Up Passed! âœ“")
````

## Problem 1: Move Zeroes â­ Easy
**Task**: Move all `0`'s to the end while maintaining the relative order of non-zero elements.

**Key Difference from Remove**: We keep zeros, just move them to end.

````python
def moveZeroes(nums: list[int]) -> None:
    """
    Algorithm:
    1. Use write pointer for next non-zero position
    2. Scan with read pointer
    3. When non-zero found, SWAP with write position (not just copy!)
    
    Why swap? To push zeros forward without losing them.
    """
    write = 0
    
    for read in range(len(nums)):
        if nums[read] != 0:
            # TODO: Swap nums[write] and nums[read]
            # Optimization: only swap if positions are different
            if write != read:
                nums[write], nums[read] = nums[read], nums[write]
            
            # TODO: Move write pointer
            write += 1
````

````python
# Test Cases
test_1 = [0,1,0,3,12]
moveZeroes(test_1)
assert test_1 == [1,3,12,0,0], f"Expected [1,3,12,0,0], got {test_1}"

test_2 = [0]
moveZeroes(test_2)
assert test_2 == [0]

test_3 = [1, 0, 2, 0, 3]
moveZeroes(test_3)
assert test_3 == [1, 2, 3, 0, 0]

print("Move Zeroes Tests Passed! âœ“")
````

### ðŸ’¡ Reflection
- Why do we SWAP instead of just copying?
- What happens if we don't check `write != read`?

## Problem 2: Remove Duplicates from Sorted Array â­ Easy
**Task**: Given a sorted array, remove duplicates in-place. Return new length.

**Key Insight**: Array is SORTED, so duplicates are adjacent!

````python
def removeDuplicates(nums: list[int]) -> int:
    """
    Algorithm:
    1. First element is always unique
    2. Compare each element with PREVIOUS
    3. If different, it's a new unique value
    """
    if not nums:
        return 0
    
    write = 1  # First element is always kept
    
    for read in range(1, len(nums)):
        # TODO: If current != previous, it's unique
        if nums[read] != nums[read - 1]:
            # TODO: Copy to write position
            nums[write] = nums[read]
            write += 1
            
    return write
````

````python
# Test Cases
test_dup = [1,1,2]
k = removeDuplicates(test_dup)
assert k == 2 and test_dup[:k] == [1,2]

test_dup2 = [0,0,1,1,1,2,2,3,3,4]
k2 = removeDuplicates(test_dup2)
assert k2 == 5 and test_dup2[:k2] == [0,1,2,3,4]

print("Remove Duplicates Tests Passed! âœ“")
````

## ðŸŽ¯ Practice: Remove Duplicates II (At Most 2)
**Challenge**: Allow at most 2 duplicates instead of 1.

````python
def removeDuplicatesII(nums: list[int]) -> int:
    """Allow at most 2 duplicates."""
    if len(nums) <= 2:
        return len(nums)
    
    write = 2  # First two elements always kept
    
    for read in range(2, len(nums)):
        # TODO: Compare with element 2 positions back
        # If different, we can include it
        if nums[read] != nums[write - 2]:
            nums[write] = nums[read]
            write += 1
    
    return write

# Test
arr = [1,1,1,2,2,3]
k = removeDuplicatesII(arr)
assert k == 5 and arr[:k] == [1,1,2,2,3]
print("Remove Duplicates II Passed! âœ“")
````

## Problem 3: Sort Colors (Dutch National Flag) â­â­ Medium
**Task**: Sort array containing only 0, 1, 2 in-place. One pass, O(1) space.

**The Three Pointer Approach:**
```
[0,0,0,...| 1,1,1,... | unseen... | 2,2,2,...]
          ^           ^          ^
         low         mid       high
```

**Rules:**
- `[0, low)`: All 0s
- `[low, mid)`: All 1s
- `[mid, high]`: Unseen
- `(high, n)`: All 2s

````python
def sortColors(nums: list[int]) -> None:
    """
    Algorithm:
    - If nums[mid] == 0: swap with low, increment both
    - If nums[mid] == 1: just increment mid
    - If nums[mid] == 2: swap with high, decrement high (don't move mid!)
    
    CRITICAL: When swapping with high, we DON'T increment mid because
    the swapped value is unknown and needs inspection!
    """
    low = 0
    mid = 0
    high = len(nums) - 1
    
    while mid <= high:
        if nums[mid] == 0:
            # TODO: Swap with low, move both low and mid
            nums[low], nums[mid] = nums[mid], nums[low]
            low += 1
            mid += 1
            
        elif nums[mid] == 1:
            # TODO: 1 is in correct region, just move mid
            mid += 1
            
        else:  # nums[mid] == 2
            # TODO: Swap with high, only decrement high
            # WHY not increment mid? We need to check what was swapped!
            nums[mid], nums[high] = nums[high], nums[mid]
            high -= 1
````

````python
# Test Cases
colors = [2,0,2,1,1,0]
sortColors(colors)
assert colors == [0,0,1,1,2,2], f"Got {colors}"

colors2 = [2,0,1]
sortColors(colors2)
assert colors2 == [0,1,2]

print("Sort Colors Tests Passed! âœ“")
````

### ðŸ’¡ Key Insight
**Why don't we increment mid after swapping with high?**

Because the value that came from `high` could be 0, 1, or 2 - we haven't seen it yet!

## ðŸ“ Summary

| Problem | Pointers | Key Action |
|---------|----------|------------|
| Move Zeroes | write, read | Swap non-zeros forward |
| Remove Dups | write, read | Copy unique values |
| Sort Colors | low, mid, high | 3-way partition |

**Key Patterns:**
1. **Write/Read**: One writes, one reads - O(N) single pass
2. **Three Pointers**: For 3-way partitioning
